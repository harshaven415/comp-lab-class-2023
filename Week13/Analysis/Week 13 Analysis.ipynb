```python
import numpy as np
import matplotlib.pyplot as mplt
import seaborn as sns
import pandas as pd
```

```python
#First, we will link to our NVT and NPT data

data_nvt = np.loadtxt('/home/hv454/comp-lab-class-2023/Week13/Outputs/NVT-1.ener')

data = pd.DataFrame(data_nvt, columns=['Step Nr.', 'Time[fs]', 'Kin.[a.u.]', 'Temp[K]', 'Pot.[a.u.]', 'Cons Qty[a.u.]', 'UsedTime[s]'])
```

```python
#First, we will present the NVT data as a table below 

data
```

```python
#Now, we will create our plot of Potential Energy vs. Time

time = data['Time[fs]']
potential = data['Pot.[a.u.]']

mplt.figure(figsize=(20,15))
mplt.plot(time, potential)
mplt.title('Potential Energy vs. Time')
mplt.xlabel('Time')
mplt.ylabel('Potential Energy [a.u.]')
mplt.grid(True)

mplt.show()
```
