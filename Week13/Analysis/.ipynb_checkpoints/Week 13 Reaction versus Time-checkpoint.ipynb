### This Jupyter notebook will determine the reaction coordinate vs time from the QM/MM data

```python
import numpy as np
import matplotlib.pyplot as mplt
import seaborn as sns
import pandas as pd
```

```python
#Similarly to the potential energy, the first step is to load in our data sets

data_coordinate = np.loadtxt('/home/hv454/comp-lab-class-2023/Week13/Outputs/monitor/MONITOR-COLVAR.metadynLog')

coordinate_data = pd.DataFrame(data_coordinate, columns=['Time [fs]', 'Reaction Coordinate', 'Position', 'X', 'Y', 'Z'])
```

```python
coordinate_data
```

```python
#Now, we will create our plot of Reaction Coordinate vs. Time

time = coordinate_data['Time [fs]']
coordinate = coordinate_data['Reaction Coordinate']

mplt.figure(figsize=(20,15))
mplt.plot(time, coordinate)
mplt.title('Reaction Coordinate vs. Time')
mplt.xlabel('Time [fs]')
mplt.ylabel('Reaction Coordinate')


mplt.grid(True)


mplt.show()
```
