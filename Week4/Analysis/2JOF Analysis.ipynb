### This Jupyter notebook will analyze the end to end distance and radius of gyration for 2JOF

```python
import matplotlib.pyplot as mplt
import mdtraj as mdt
```

```python
pwd
```

```python
traj = traj = mdt.load_xtc('/home/hv454/comp-lab-class-2023/Week4/Setup/2jof_amber/gromacs/md_center_nosolvent.xtc', top='topol_water-free.gro')
```

```python
#First, we will compute end to end distance and the radius of gyration

end_to_end = mdt.compute_distances(traj, [[0, traj.n_atoms-1]])

radius_of_gyration = mdt.compute_rg(traj)
```

```python
#Now, we will plot the end to end distance here

mplt.plot(end_to_end)
mplt.xlabel('Frame')
mplt.ylabel('End to end distance (nm)')
mplt.title('End to end distance of 2JOF')
```

```python
#Next, we will plot the radius of gyration 

mplt.plot(radius_of_gyration)
mplt.xlabel('Frame')
mplt.ylabel('Radius of Gyration (nm)')
mplt.title('Radius of Gyration of 2JOF')
```
