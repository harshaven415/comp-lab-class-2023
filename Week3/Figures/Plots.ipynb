# This Jupyter Notebook presents data plots
#### The data plots analyze the trajectory of the xvg files from the Gromacs Tutorial

```python
import numpy as np
import matplotlib.pyplot as mplt
import pandas as pnd
import seaborn as sns

#First, we are importing libraries to help with plotting
```

```python
data_density = np.loadtxt('density.xvg', skiprows=24)
data_d = pnd.DataFrame(data_density, columns = ['Time', 'Density'])
```

```python
data_d
```

```python
#Now, we will repeat the above for each of the data sets
data_gyrate = np.loadtxt('gyrate.xvg', skiprows=27)
data_g = pnd.DataFrame(data_gyrate, columns= ['Time', 'Rg', 'Rg/sX/N', 'Rg/sY/N', 'Rg/sX/N'])
```

```python
data_g
```

```python
data_potential = np.loadtxt('potential.xvg', skiprows=24)
data_pot = pnd.DataFrame(data_potential, columns= ['Time', 'Energy'])
```

```python
data_pot
```

```python
data_pressure = np.loadtxt('pressure.xvg', skiprows=24)
data_pre = pnd.DataFrame(data_pressure, columns = ['Time', 'Pressure'])
```

```python
data_pre
```

```python
data_rmsd_xtal = np.loadtxt('rsmd_xtal.xvg', skiprows = 18)
data_rmxt = pnd.DataFrame(data_rmsd_xtal, columns = ['Time', 'RMSD'])
```

```python
data_rmxt
```

```python
data_temp = np.loadtxt('temperature.xvg', skiprows= 24)
data_t = pnd.DataFrame(data_temp, columns=['Time', "Temperature"])
```

```python
data_t
```

```python
#Now, all of the xvg files have been made into readable tables
#The next step is to plot them below
```

```python
#Similar to the first assignment, we will make 6 plots of data
#However, we have to identify different data sets

fig, axes = mplt.subplots(2,3, figsize = (25,25))
sns.lineplot(data=data_d, x="Time", y="Density", ax=axes[0,0])
sns.lineplot(data=data_g, x="Time", y="Rg", ax=axes[0,1])
sns.lineplot(data=data_pot, x="Time", y="Energy", ax=axes[0,2])
sns.lineplot(data=data_pre, x="Time", y="Pressure", ax=axes[1,0])
sns.lineplot(data=data_rmxt, x="Time", y="RMSD", ax=axes[1,1])
sns.lineplot(data=data_t, x='Time', y='Temperature', ax=axes[1,2])
```
