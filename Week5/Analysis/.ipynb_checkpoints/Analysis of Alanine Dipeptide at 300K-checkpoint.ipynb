This Jupyter Notebook will compute the phi and psi dihedral angles of alanine dipeptide, and will make a free energy surface as well

```python
import matplotlib.pyplot as mplt
import numpy as np
import mdtraj as mdt
```

```python
#First, the trajectories will be loaded 

traj = mdt.load_trr('../Inputs/md_adp.trr', top='../Inputs/md_adp.gro')
```

```python
#Next, the phi and psi angles will be computed 

phi = mdt.compute_phi(traj)
psi = mdt.compute_psi(traj)
```

```python
#Next, the phi and psi angles will be plotted

mplt.plot(phi[1][:,0], label = "Phi (Degrees)")
mplt.plot(psi[1][:,0], label = "Psi (Degrees)")

#Now we will add axis labels 

mplt.xlabel('Phi')
mplt.ylabel('Psi')

mplt.show()
```

```python
#Next, we will compute a 2d Histogram of the Phi and Psi Angles 
H, x_edges, y_edges = np.histogram2d(phi[1][:,0], psi[1][:,0], bins = 100)

#Finally, we will plot the histogram

mplt.imshow(H, interpolation = 'nearest', origin = 'lower', extent = [x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]])

mplt.xlabel('Phi')
mplt.ylabel('Psi')
mplt.colorbar()
mplt.show()
```

```python
#Now we will remove the background coloring and provide a title 

mplt.imshow(-np.log(H), interpolation='nearest', origin='lower', extent=[x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]])

mplt.xlabel('Phi')
mplt.ylabel('Psi')
mplt.title('2D Histogram of Phi and Psi Angles')
mplt.colorbar()
mplt.show()
```
