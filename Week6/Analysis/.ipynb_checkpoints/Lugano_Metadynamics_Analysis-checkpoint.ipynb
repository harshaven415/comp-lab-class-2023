### This Jupyter notebook will plot the 2D Metadynamics Countour plots from the Lugano tutorial

```python
import matplotlib.pyplot as mplt
import numpy as np
import mdtraj as mdt
```

```python
data_A = np.loadtxt('fes_A.dat', comments=['#'])
data_B = np.loadtxt('fes_B.dat', comments=['#'])
```

```python
#First, we willl shape the data as needed

#reshape the data
phi_new_A = data_A[:33,0]
psi_new_A = data_A[::33,1]
fes_A = data_A[:,2]
fes_A_new = fes_A.reshape(4,9)


phi_new_B = data_B[:33,0]
psi_new_B = data_B[::33,1]
fes_B = data_B[:,2]
fes_B_new = fes_B.reshape(4,9)

```

```python
print(fes_A.shape)
```

```python
#Now, we will plot a countour of A and B using subplots

fig, (ax1, ax2) = mplt.subplots(1, 2, figsize=(10,5))
ax1.contourf(phi_new_A, psi_new_A, fes_A_new, 100, cmap='viridis')
ax1.set_xlabel(r'$\phi$')
ax1.set_ylabel(r'$\psi$')
ax1.set_title('A')
ax2.contour(phi_new_B, psi_new_B, fes_B_new, 100, cmap='viridis')
ax2.set_xlabel(r'$\phi$')
ax2.set_ylabel(r'$\psi$')
ax2.set_title('B')
mplt.show()
```
