```python
import matplotlib.pyplot as mplt
import numpy as np
import mdtraj as mdt
```

```python
#First, we will load in the data from the COLVAR files

data_A = np.loadtxt("/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/topolA/diala/COLVAR")
data_B = np.loadtxt("/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/topolB/diala/COLVAR")
```

```python
#Next, we will import the trajectories

traj_A = mdt.load("/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/topolA/diala/traj_comp.xtc", top="/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/diala/dialaA.pdb")
traj_B = mdt.load("/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/topolA/diala/traj_comp.xtc", top="/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/diala/dialaB.pdb")
```

```python
#Next, we will now compute the phi and psi angles from the trajectories

phi_A = mdt.compute_phi(traj_A)
psi_A = mdt.compute_psi(traj_A)
phi_B = mdt.compute_phi(traj_B)
psi_B = mdt.compute_psi(traj_B)
```

```python
#Now, using the data above, we can plot A and B
mplt.plot(data_A[:,1], data_A[:,2], 'o', label="A")
mplt.plot(data_B[:,1], data_B[:,2], 'o', label="B")
mplt.xlabel("Phi (radians)")
mplt.ylabel("Psi (radians")
mplt.legend()
mplt.show()
```

```python
#Next, we can try plotting phi vs psi on the same plot as well
mplt.plot(phi_A[1], psi_A[1], 'o', label='A')
mplt.plot(phi_B[1], psi_B[1], 'o', label='B')
mplt.xlabel('Phi (radians')
mplt.ylabel('Psi (radians')
mplt.legend()
mplt.show()
```

```python
#Finally, I will plot the MD Traj data
mplt.plot(phi_A[1], psi_A[1], 'o', label='A')
mplt.plot(phi_B[1], psi_B[1], 'o', label='B')
mplt.xlabel('Phi (radians')
mplt.ylabel('Psi (radians')
mplt.legend()
mplt.show()
```

```python
#Using the above data, we can get the standard deviations as required

#compute standard deviation of phi and psi for A and B
std_phi_A = np.std(data_A[:,1])
std_psi_A = np.std(data_A[:,2])
std_phi_B = np.std(data_B[:,1])
std_psi_B = np.std(data_B[:,2])
#print out standard deviation
print("Standard deviation of phi for A is", std_phi_A)
print("Standard deviation of psi for A is", std_psi_A)
print("Standard deviation of phi for B is", std_phi_B)
print("Standard deviation of psi for B is", std_psi_B)
```
