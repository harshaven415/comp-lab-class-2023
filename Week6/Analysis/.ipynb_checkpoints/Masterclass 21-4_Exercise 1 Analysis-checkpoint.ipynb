```python
#Firsty, we will import the trajectories

traj_A = mdt.load('/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/run_A/traj_comp.xtc', top='/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/data/diala/dialaA.pdb')
traj_B =mdt.load('/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/run_B/traj_comp.xtc', top='/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/data/diala/dialaB.pdb')
```

```python
#Next, we will compute the phi and psi from the trajectories for A and B

phi_A = mdt.compute_phi(traj_A)
psi_A = mdt.compute_psi(traj_A)
phi_B = mdt.compute_phi(traj_B)
psi_B = mdt.compute_psi(traj_B)
```

```python
data_A = np.loadtxt('/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/run_A/COLVAR')
data_B = np.loadtxt('/home/hv454/comp-lab-class-2023/Week6/masterclass-21-4/Exercise_1/run_B/COLVAR')
```

```python
phi_A
```

```python
#First, we will plot phi and psi using the data from COLVAR

mplt.plot(data_A[:,1], data_A[:,2], 'o', label="A")
mplt.plot(data_B[:,1], data_B[:,2], 'o', label="B")
mplt.xlabel("Phi (radians)")
mplt.ylabel("Psi (radians")
mplt.legend()
mplt.show()
```

```python
#Next, we will plot the data from phi vs psi from A and B on the same plot

mplt.plot(phi_A[1], psi_A[1], 'o', label='A')
mplt.plot(phi_B[1], psi_B[1], 'o', label='B')
mplt.xlabel('Phi (radians')
mplt.ylabel('Psi (radians')
mplt.legend()
mplt.show()
```

```python
#Finally, we will determine the standard deviation for phi and psi for A and B

std_phi_A = np.std(data_A[:,1])
std_psi_A = np.std(data_A[:,2])
std_phi_B = np.std(data_B[:,1])
std_psi_B = np.std(data_B[:,2])


print("Standard deviation of phi for A is", std_phi_A)
print("Standard deviation of psi for A is", std_psi_A)
print("Standard deviation of phi for B is", std_phi_B)
print("Standard deviation of psi for B is", std_psi_B)
```
