### This Jupyter notebook will plot the energy vs phi from Sim_A and Sim_B

```python
import matplotlib.pyplot as mplt
import numpy as np
import mdtraj as mdt
```

```python
#First, we will load the grid files for A and B

data_A = np.loadtxt('metad.grid_A.dat',comments=['@','#'])
data_B = np.loadtxt('metad.grid_B.dat',comments=['@','#'])
```

```python
#Next, we will plot the phi against the energy

mplt.plot(data_A[:,0],data_A[:,1], label = "Sim_A")
mplt.plot(data_B[:,0],data_B[:,1], label = "Sim_B")

mplt.xlabel('Phi (Radians)')
mplt.ylabel('Energy (kJ/mol)')
mplt.title('Energy vs Phi (Radians)')

mplt.legend()
mplt.show()
```
